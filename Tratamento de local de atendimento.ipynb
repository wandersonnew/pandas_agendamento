# Tratamento de dados

## importando as bibliotecas

```python
import pandas as pd
import numpy as np
import requests
from openpyxl import Workbook
wb = Workbook()

from requests.packages.urllib3.exceptions import InsecureRequestWarning

# Desabilitar os avisos de solicitação insegura
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)
```

### carregando os dados de um arquivo CSV e verificando as cinco primeiras linhas

```python
dados = pd.read_csv('local_atendimento.csv')
dados.head()
```

### Tratando dados vazios e convertendo para um valor padrão se vazio

```python
dados['nome_fantasia'] = dados['nome_fantasia'].fillna('Não informado')
dados['nome_fantasia'] = dados['nome_fantasia'].astype(str)

dados['fone'] = dados['fone'].fillna('Não informado')
dados['fone'] = dados['fone'].astype(str)

dados['cep'] = dados['cep'].fillna(00000000)
dados['cep'] = dados['cep'].astype(int)

dados.head()
```

### Gerando arquivo XLSX

```python
dados.to_excel('dados_local_atendimento.xlsx', engine='xlsxwriter')
```
